## Calculate the average number of runs scored in each inning.

```python
cursor.execute('''
With temp_table as (select ID
from ball 
group by ID, BattingTeam
order by sum(total_run) desc limit 1)

select t1.ID, City, Date, Season,Team1,Team2, WinningTeam, total_run, BattingTeam

from ipl.match t1 join ipl.ball t2 on t1.id = t2.id   

where t1.ID IN (select * from temp_table) and BattingTeam = WinningTeam;

''')
```

```python
pd.DataFrame(cursor.fetchall(), columns=["ID", "City", "Date", "Season", "Team1", "Team2", "WinningTeam", "run", "Batting Team"])
```

## Find the team that scored the highest total runs in a single match.

```python
cursor.execute('''
With temp_table as (select ID
from ball 
group by ID, BattingTeam
order by sum(total_run) desc limit 1)

select t1.ID, City, Date, Season,Team1,Team2, WinningTeam, sum(total_run) 

from ipl.match t1 join ipl.ball t2 on t1.id = t2.id   

where t1.ID IN (select * from temp_table) and BattingTeam = WinningTeam;

''')
```

## Find the matches where a player received the "Player of the Match" award and also ended up being the top scorer

```python
cursor.execute("""
WITH temp_table AS
    (
        SELECT t1.Player_of_Match, t2.batter, t2.runs, t2.ID 
        FROM ipl.match t1 left join 
        (
            SELECT ID, batter, sum(total_run) as runs
            FROM ipl.ball
            Group BY ID, batter
        ) t2 
        on t1.ID = t2.ID
    )

SELECT t3.*
FROM (SELECT * from temp_table) t3 JOIN (
                                            select ID, max(runs) as max_runs 
                                            FROM (select * from temp_table) t1
                                            GROUP BY t1.ID
                                        )   t2
on t2.ID = t3.ID
WHERE runs = max_runs and Player_of_Match = batter;
""")
```

```python
pd.DataFrame(cursor.fetchall(), columns=["Player_of_Match", "Batter", "Runs", "Match-ID"])
```

## Find the matches where the total number of extras (extra_runs) is greater than 10.

```python
cursor.execute("""
select t1.ID, Season, MatchNumber ,Team1, Team2, Extra_Run from ipl.match t1 left join (SELECT ID, sum(extras_run) as Extra_Run from ipl.ball
group by ID having Extra_Run > 10) t2 on t1.ID = t2.ID 
""")
```

```python
pd.DataFrame(cursor.fetchall(), columns=["ID", "Season", "MatchNumber", "Team 1", "Team 2", 'Extra Run'])
```

## Find the team with the highest average run rate across all matches.

```python
cursor.execute('''
SELECT round((sum(total_run)/count(*)) * 6, 2) as RunRate, BattingTeam, t1.ID, Team1, Team2, Venue, Date, Season
FROM ipl.ball t1 join ipl.match t2 on t1.ID = t2.ID
GROUP BY t1.ID, BattingTeam 
order by RunRate desc limit 1
''')
```

```python
cursor.fetchall()
```

## Retrieve the details of matches where a Viral Kohli  was the highest scorer.

```python
cursor.execute('''
WITH common_table as 
(select ID, batter, sum(batsman_run) as Runs
from ipl.ball
group by ID, batter)

select ID, Runs, batter, Team1, Team2, WinningTeam, Venue, Date from 

(common_table) natural join  ipl.match 

where (ID , Runs) in 

(select ID, max(Runs) as runs from 

(common_table) 

group by ID) and batter = "V Kohli";
''')
```

```python
pd.DataFrame(cursor.fetchall(), columns=["Match ID", "Runs", "batter", "Team1", "Team2", "WinningTeam", "Venue", "Date"])
```
